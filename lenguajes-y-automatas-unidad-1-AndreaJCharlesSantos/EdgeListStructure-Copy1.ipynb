# Edge List Structure - Estructura de Lista de Aristas

Realice lo siguiente:
  * Programe la representación de grafos mediante un la **Estructura de Lista de Aristas**.
  * Mediante networkx muestre el resultado gráfico del gráfo.

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def AddNode(G,Node):
    for i in node:
        G.add_node(i)
    return G

def AddEdge(Edge):
    Gr= nx.Graph()
    for k in Edge:
        Gr.add_edge(*k)
    return Gr

def ImprimirGrafo(G):
    nx.draw(G, with_labels=True, font_weight='bold', node_size=1500, node_color='pink', font_size=20)
```

```python
G = nx.Graph()
node_list = {'a', 'b', 'c', 'd', 'e'}
edge_list = {('a', 'b'), ('b', 'd'), ('a', 'c'), ('c', 'e'), ('e', 'd'),('c','b')}
AddNode(G, node_list)
G = AddEdge(edge_list)
print("Nodes from G:", list(G.nodes))
print("Edges from G:", list(G.edges))
ImprimirGrafo(G)
```
