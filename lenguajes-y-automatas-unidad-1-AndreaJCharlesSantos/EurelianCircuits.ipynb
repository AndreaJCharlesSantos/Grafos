# Circuitos Eurelianos

Realice lo siguiente:

  * Dado un grafo de entrada, se deberÃ¡ responder lo siguiente:
    * El grafo es un grafo Eureliano o un grafo no-Eureliano
    * Determinar el circuito Eureliano
  * Mediante networkx mostrar el grafo con su circuito Eureliano.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def Comprobate(graph):
    compr = False
    for i in graph.nodes():
        if(len(graph[i]) % 2 == 0):
            compr = True
        else:
            compr = False
            break
    return compr

def find_path(graph):
    eur = []
    ACTUAL = graph[0][0]
    eur.append(ACTUAL)
    while len(graph) > 0:
        for edge in graph:
            if ACTUAL in edge:
                if edge[0] == ACTUAL:
                    ACTUAL = edge[1]
                else:
                    ACTUAL = edge[0]
                graph.remove(edge)
                eur.append(ACTUAL)
    return eur
```

#Ejemplo de uno eureliano

```python
graph2 = [("A", "B"), ("A","D"), ("B", "C"), ("C", "D")]
nodes2 = ["A","B","C","D"]
G2 = nx.Graph()
G2.add_nodes_from(nodes2)
G2.add_edges_from(graph2)
res = Comprobate(G2)
if res == True:
    print("Del A: ",find_path(graph2))
else:
    print("No es eureliano nimodo")
nx.draw(G2, with_labels=True, font_weight='bold', node_size=1500, node_color='pink', font_size=20)
```

# Ejemplo cuando no es eureliano

```python
graph3 = [("A", "B"), ("A","D"), ("B", "C"), ("C", "D"), ("A","C")]
nodes3 = ["A","B","C","D"]
G3 = nx.Graph()
G3.add_nodes_from(nodes3)
G3.add_edges_from(graph3)
res = Comprobate(G3)
if res == True:
    print(find_path(graph3))
else:
    print("No es eureliano nimodo")
nx.draw(G3, with_labels=True, font_weight='bold', node_size=1500, node_color='pink', font_size=20)
```
